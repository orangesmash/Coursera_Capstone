```python
results = requests.get(url).json()
results
```

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']
venues
```

```python
# tranform venues into a dataframe
dataframe = pd.json_normalize(venues)
dataframe.head()
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
df_sg = dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
df_sg['categories'] = df_sg.apply(get_category_type, axis=1)

# clean column names by keeping only last term
df_sg.columns = [column.split('.')[-1] for column in df_sg.columns]

df_sg
```

```python
venues_map = folium.Map(location=[latitude_sg, longitude_sg], zoom_start=13) # generate map centred around central Singapore

# add a red circle marker to represent central Singapore
folium.CircleMarker(
    [latitude_sg, longitude_sg],
    radius=10,
    color='red',
    popup='Conrad Hotel',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map)

# add the Halal restaurants as blue circle markers
for lat, lng, label in zip(df_sg.lat, df_sg.lng, df_sg.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)

# display map
venues_map
```

```python
#Setting the cluster number as number of labels
kclusters=len(df_sg.Label.unique())

# create map
map_sg = folium.Map(location=[latitude_sg, longitude_sg], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]


# add markers to the map
markers_colors = []
for categories in zip(df_sg['Label']):
    label = '{}'.format(categories)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [latitude_sg, longitude_sg],
        radius=5,
        popup=label,
        color=rainbow[categories],
        fill=True,
        fill_color=rainbow[categories-1],
        fill_opacity=0.7).add_to(map_sg)
```
